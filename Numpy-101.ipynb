```python
import numpy as np
print(np.__version__)

arr = np.arange(10)  # 一维数组 0-9
print(arr)

arr = np.full((3, 3), True, dtype=bool)  # 3x3 True矩阵
arr2 = np.ones((3, 3), dtype=bool)  # 同上
print(arr, arr2)

# 从数组中按条件提取
arr = np.arange(10)
print(arr[arr%2 == 1], arr)  # 取出所有奇数

# 按条件替换
arr[arr%2 == 1] = -1  # 将所有奇数替换为-1
print(arr)
```

```python
# 按替换但不覆盖原属组
arr = np.arange(10)
out = np.where(arr%2 == 1, -1, arr)  # 利用where函数建立副本
print(arr, out)
print('----------------')

# 将一维数组重组为二维矩阵
arr2 = arr.reshape(2, -1)  # -1自动安排列
print(arr2)
print('----------------')

# 垂直堆叠
a = np.arange(10).reshape(2, -1)
b = np.repeat(1, 10).reshape(2, -1)
print(a, b)
print(np.concatenate([a, b], axis=0))  # concatenate函数按照指定轴拼接
print(np.vstack([a, b]))  # vstack函数垂直堆叠
print(np.r_[a, b])  # r_函数
print('----------------')

# 水平堆叠
print(np.concatenate([a, b], axis=1))  # concatenate函数按照指定轴拼接
print(np.hstack([a, b]))  # hstack函数水平堆叠
print(np.c_[a, b])  # c_函数
print('----------------')

# 自定义构建序列
a = np.array([1,2,3])
b = np.r_[np.repeat(a, 3), np.tile(a, 3)]  # repease 每个元素重复，tile整体重复
print(a)
print(b)
print('----------------')

# 获得两个数组的公共元素
a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])
print(np.intersect1d(a, b))
print('----------------')

# 从一个数组中去掉与另一个数组重复的元素
a = np.array([1,2,3,4,5])
b = np.array([5,6,7,8,9])
print(np.setdiff1d(a,b))
print(a)
```
